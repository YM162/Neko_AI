```python
from statistics import mean
import random
Datasetinput=[]
Datasetoutput=[]
dataclose=data["close"]
datavolume=data["volume"]
dataclose.dropna(how='any',inplace=True)
datacloserel=[]
for i in range(1,len(dataclose)):
    datacloserel.append(((dataclose[i]/dataclose[i-1])-1)*100)

for i in reversed(range(130,len(datacloserel))):
    t=0
    tempimput=[]
    while t < 10:
        tempimput.append(datacloserel[i-t])
        t=t+1
    Datasetoutput.append(datacloserel[i-t])
    #tempimput.append(datavolume[i-t-1]/mean([datavolume[i-t-2],datavolume[i-t-3],datavolume[i-t-4],datavolume[i-t-5]]))
    Datasetinput.append(tempimput)

mapIndexPosition = list(zip(Datasetinput, Datasetoutput))
random.shuffle(mapIndexPosition)
Datasetinput, Datasetoutput = zip(*mapIndexPosition)
```
