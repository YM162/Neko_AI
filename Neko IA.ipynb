```python
import tensorflow as tf
import keras

from keras.optimizers import SGD
from keras.layers import MaxPooling1D
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import LSTM
from keras import layers
from keras.layers import Conv1D
from keras.layers.embeddings import Embedding
from keras.layers import BatchNormalization
from keras.layers.merge import concatenate

inputs = keras.Input(shape=(50,2), name="inputs")
x= layers.BatchNormalization()(inputs)

x= layers.Conv1D(filters=800, kernel_size=5, padding='same', activation='relu',kernel_initializer="normal")(x)
x= layers.Conv1D(filters=800, kernel_size=5, padding='same', activation='relu',kernel_initializer="normal")(x)
y= layers.MaxPooling1D(pool_size=2,padding='same')(x)
#y= layers.BatchNormalization()(x)

x= layers.Conv1D(filters=800, kernel_size=5, padding='same', activation='relu',kernel_initializer="normal")(y)
x= layers.Conv1D(filters=800, kernel_size=5, padding='same', activation='relu',kernel_initializer="normal")(x)
#x= layers.MaxPooling1D(pool_size=2,padding='same')(x)
#x= layers.BatchNormalization()(x)

z= layers.Add()([x,y])
z2= layers.Flatten()(z)
z2= layers.Dense(100)(z2)

x= layers.Conv1D(filters=800, kernel_size=5, padding='same', activation='relu',kernel_initializer="normal")(z)
x= layers.Conv1D(filters=800, kernel_size=5, padding='same', activation='relu',kernel_initializer="normal")(x)
x= layers.MaxPooling1D(pool_size=2,padding='same')(x)


x= layers.LSTM(300,kernel_initializer="normal",activation='softmax',return_sequences = True)(x)
x= layers.LSTM(100,kernel_initializer="normal",activation='softmax')(x)

x= layers.Flatten()(x)
z= layers.Add()([x,z2])


x= layers.Dense(300,kernel_initializer="normal",activation='relu')(z)
x= layers.Dense(100,kernel_initializer="normal",activation='relu')(x)
x= layers.Dense(50,kernel_initializer="normal",activation='relu')(x)

outputs = layers.Dense(1,name="outputs",activation="sigmoid")(x)
model = keras.Model(inputs=[inputs],outputs=[outputs])

keras.utils.plot_model(model, "multi_input_and_output_model.png", show_shapes=True)




model.compile(optimizer=keras.optimizers.Adam(lr=1e-6), loss="binary_crossentropy",metrics=["accuracy"])
#model.compile(optimizer="adam", loss='mean_absolute_error',metrics=["accuracy"])
#model.save_weights("model1Conv2LSTM2Dense.h5")
```

```python
import numpy as np
import tensorflow as tf
import keras
from keras.optimizers import SGD
from keras.layers import MaxPooling1D
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import LSTM
from keras.layers import Conv1D
from keras.layers.embeddings import Embedding

model = tf.keras.Sequential()
model.add(Conv1D(filters=32, kernel_size=3, padding='same', activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.5))
model.add(LSTM(300,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dropout(0.2))
model.add(tf.keras.layers.Dense(1000))
model.add(tf.keras.layers.Dense(100))
model.add(tf.keras.layers.Dense(1,activation="sigmoid"))
model.compile(optimizer="adam", loss='binary_crossentropy',metrics=["accuracy"])
model.build(input_shape=(None,10, 1))
model.load_weights("model1Conv2LSTM2Dense1.h5")

def predecir(prerealdata,cut):


    #realdata tiene los últimos 11 datos de close, en intervalos de 5m
    #realdata=(100,110,103,109,105,102,101,102,104,90,80)
    realdata=list( prerealdata[i] for i in [0,5,10,15,20,25,30,35,40,45,50] )
    
    realdatarel=[]

    for i in range(1,len(realdata)):
        realdatarel.append(((realdata[i]/realdata[i-1])-1)*100)
    
    realdatarel_array = np.array(realdatarel)
    ncols = realdatarel_array.shape[0]
    realdatarel_array = realdatarel_array.reshape(1, ncols, 1)
    #modelo red

    #predecir y aplicar el cut para el return
    realdata_Y = model.predict(realdatarel_array)
    #plt.plot(realdata)
    #plt.show()
    print(realdata_Y)
    if realdata_Y < cut:
        return 0 #comprar short
    if realdata_Y > 1-cut:
        return 1 #comprar long
    else:
        return -1 #no está seguro
predecir(range(1,55),0.5)
```
